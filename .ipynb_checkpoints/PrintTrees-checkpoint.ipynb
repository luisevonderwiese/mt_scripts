```python
from Bio import Phylo
import os
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc('font', size=6)
os.mkdir('../output/drawings')
for prefix in ["COGNATE", "MORPHOLOGICAL"]:
    for s in ["T_3", "T_4", "T_5"]:
        tree = Phylo.read('../output/raxml/' + prefix + "/" + s + ".raxml.bestTree", "newick")
        tree.ladderize()
        #print(tree)
        fig = plt.figure(figsize=(10, 10), dpi=100)
        axes = fig.add_subplot(1, 1, 1)
        axes.set_title(prefix + " " + s)
        #Phylo.draw(tree, axes=axes)
        Phylo.draw(tree, axes=axes, do_show=False)
        plt.savefig('../output/drawings/' + prefix + " " + s + '.png', dpi=fig.dpi)
```
