```python
import os
import matplotlib.pyplot as plt
import numpy as np
with open("../data/bootstrap/morpho_filtered_large_step10000000.raxml.log", 'r') as log_file:
    lines = log_file.readlines()
    lhs = []
    for line in lines:
        data = line.split(" ")
        if len(data) < 8:
            continue
        if not data[6] == "logLikelihood:":
            continue
        lhs.append(float(data[7]))
    plt.scatter(range(len(lhs)), lhs, s=8)
    plt.xlabel('bootstrap trees')
    plt.ylabel('logLikelihood')
    plt.show()
    print(np.var(lhs))
    
```
