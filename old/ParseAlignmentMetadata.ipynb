```python
import os
d = "../database/datasets/language_families/"
with os.scandir(d) as it:
    for entry in it:
        f = ""
        if entry.name.endswith(".phy") and entry.is_file():
            f = "phylip"
        else:
            if entry.name.endswith(".nex") and entry.is_file():
                f = "nexus"
            else:
                continue
        data = entry.name.split('.')
        if data[1] == 'cc':
            t = "cognate"
        if data[1] == 'pmi':
            t = "pmi"
        if data[1] == 'ss_sc':
            t = "cognate+soundclass"
        if data[1] == 'sc':
            t = "soundclass"
        name = '.'.join(data[:-1])
        
        
        with open(d+ entry.name, 'r') as file:
            lines = file.readlines()
            if f == "nexus":
                data = lines[2].split(" ")
                taxa = data[1].split("=")[1][:-2]
                sites = "?"
            if f == "phylip":
                data = lines[0].split(" ")
                taxa = data[0]
                sites = data[1][:-1]
        print(name + "\t" + taxa + "\t" + sites + "\t" + t + "\t" + f + "\t" + "Gerhard Jäger")
```
